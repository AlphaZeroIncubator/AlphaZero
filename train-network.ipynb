```python
import alphazero

model = alphazero.PhilipNet(alphazero.Connect4(), in_channels=3)


import torch

optimizer = torch.optim.Adam(model.parameters())

device = 'cuda:0'

model = alphazero.training.train(alphazero.Connect4,
                                 model,
                                 optimizer,
                                 alphazero.TicTacToeConverter,
                                 n_games=2,
                                 n_mcts_iter=5,
                                 subset_size=2,
                                 temperature=1,
                                 device=device,
                                 )
```

```python
model = model.to('cpu')
```
